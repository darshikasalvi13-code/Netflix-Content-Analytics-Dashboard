# 🧹 Data Cleaning

Real-world datasets often contain missing values, inconsistent formats, duplicate records, and incorrect data types.

The objective of this step is to clean and prepare the dataset to ensure accurate analysis and reliable business insights.

The following cleaning tasks will be performed:

- Handle missing values
- Remove duplicate records
- Convert data types
- Standardize text values where required
- Verify the cleaned dataset

```python
df_clean = df.copy()
```

```python
df_clean.head(2)
```

```python
df_clean.isnull().sum()
```

```python
missing_percentage = (df_clean.isnull().sum() / len(df_clean)) * 100
missing_percentage.sort_values(ascending = False)
```

```python
df_clean['director'] = df_clean['director'].fillna('Unknown')
df_clean['director'].head(3)
```

```python
df_clean['cast'] = df_clean['cast'].fillna('Unknown')
df_clean['cast'].tail()
```

```python
df_clean['country'] = df_clean['country'].fillna('Unknown')
df_clean['country'].head()
```

```python
df_clean['rating'] = df_clean['rating'].fillna('Not Rated')
df_clean['rating'].head()
```

```python
df_clean = df_clean.dropna(subset=['date_added'])
```

```python
df_clean.isnull().sum()
```

```python
df_clean.duplicated().sum()
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
df_clean.duplicated().sum()
```

```python
df_clean['date_added'].head(20)
```

```python
df_clean['date_added'].dtype
```

```python
df_clean['date_added'] = pd.to_datetime(
    df_clean['date_added'],
    errors='coerce')                            #If Pandas finds an invalid date, errors='coerce' converts it to NaT (Not a Time) instead of throwing an error.
```

```python
df_clean = df_clean.dropna(subset=['date_added'])
```

```python
df_clean.info()
```

# ⚙️ Feature Engineering

Feature Engineering is the process of creating new features from existing data to improve analysis and uncover meaningful insights.

In this step, new columns will be created from the existing dataset to support trend analysis, time-based analysis, and content categorization.

The following features will be created:

- Year Added
- Month Added
- Content Age
- Duration Category

```python
df_clean['year_added'] = df_clean['date_added'].dt.year
```

```python
df_clean[['date_added','year_added']].head()
```

```python
df_clean['month_added'] = df_clean['date_added'].dt.month_name()
```

```python
df_clean[['date_added','month_added']].tail()
```

```python
current_year = 2026
df_clean['content_age'] = current_year - df_clean['release_year']
```

```python
df_clean[['release_year','content_age']].head()
```

```python
from datetime import datetime
current_year = datetime.now().year
```

```python
def duration_category(duration):
    if 'min' in str(duration):
        return 'Movie'
    elif 'Season' in str(duration):
        return 'TV Show'
    else:
        return 'Unknown'
```

```python
df_clean['duration_category'] = df_clean['duration'].apply(duration_category)
```

```python
df_clean[['duration','duration_category']].head()
```

```python
df_clean.head()
```

```python
df_clean.to_csv("netflix_titles_featured.csv", index=False)
```

## 📌 Feature Engineering Summary

The following new features were created to enhance the dataset:

- **Year_Added:** Extracted the year content was added to Netflix.
- **Month_Added:** Extracted the month content was added.
- **Content_Age:** Calculated the age of each title based on its release year.
- **Duration_Category:** Classified content as either a Movie or TV Show based on the duration field.

These engineered features improve the dataset and enable more meaningful trend analysis and visualizations.

## 📚 Key Learnings

- Feature Engineering enhances the value of a dataset.
- Time-based features help analyze trends over different years and months.
- Creating meaningful categories simplifies analysis.
- Well-designed features improve both exploratory analysis and machine learning performance.

# 📈 Data Visualization

Data visualization is the graphical representation of information and data.

In this step, we will create various charts to better understand Netflix's content distribution, trends, and patterns.

The visualizations include:

- Movies vs TV Shows
- Top Producing Countries
- Top Genres
- Content Ratings
- Content Added by Year
- Top Directors
- Correlation Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6,5))
df_clean['type'].value_counts().plot(
    kind='bar'
)
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.show()
```

### 📌 Interpretation

The chart shows that Movies dominate Netflix's content library, indicating that the platform historically focused more on films than TV Shows.

```python
plt.figure(figsize=(10,6))

top_country.plot(
    kind='bar',
    color='crimson'
)

plt.title("Top 10 Countries by Number of Titles")
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.show()
```

## 📌 Interpretation¶
The United States contributes the highest number of titles, followed by India and the United Kingdom, highlighting Netflix's strong global presence.

```python
top_genres = df_clean['listed_in'].value_counts().head(10)
plt.figure(figsize=(12,6))

top_genres.plot(
    kind='bar',
    color='yellow'
)

plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")

plt.show()
```

```python
plt.figure(figsize=(8,5))

rating_count.head(10).plot(
    kind='bar',
    color='pink'
)

plt.title("Top Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

```python
year_added = df_clean['year_added'].value_counts().head(10)
plt.figure(figsize=(17,6))

year_added.plot(
    kind='line',
    marker='o',
    linewidth=2
)

plt.title("Netflix Content Added by Year")
plt.xlabel("Year")
plt.ylabel("Titles Added")

plt.show()
```

## 📌 Interpretation¶¶
Netflix experienced rapid growth in content additions after 2015, reflecting its global expansion and increased investment in original content.

```python
top_directors = df_clean['director'].value_counts().head(10)
plt.figure(figsize=(12,6))

top_directors.plot(
    kind='bar',
    color='green'
)

plt.title("Top Directors on Netflix")
plt.xlabel("Director")
plt.ylabel("Number of Titles")

plt.xticks(rotation=45)

plt.show()
```

```python
correlation = df_clean.select_dtypes(include='number').corr()
correlation
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    correlation,
    annot=True,
    cmap="Blues"
)

plt.title("Correlation Heatmap")

plt.show()
```

### 📌 Interpretation

The heatmap illustrates relationships between numerical features.

Positive values indicate a direct relationship, while negative values indicate an inverse relationship.

This helps identify whether numerical variables are related and supports further analysis.
